```python
import psutil
import re
import pandas as pd
```

# Display all the processes as a data-frame
---
1. create a dictionary of the processes running
2. Transpose the dataframe that is created from dictionary

```python
processes_data: dict ={p.pid: p.info for p in psutil.process_iter(["name", "username"])}
processes = pd.DataFrame(data=processes_data)\
                .transpose()\
                # .reset_index()\
                # .rename(columns={"index":"pid"})
processes
```
